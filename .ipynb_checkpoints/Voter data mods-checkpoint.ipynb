```python
df = pd.read_excel("Penn currentvotestats.xlsx")
for i in np.arange(3, 10, 2):
    df = df.drop(columns=[f"Unnamed: {i}"])
```

```python
df = df.drop(index=67)
```

```python
df["Orientation"] = df.apply(
    lambda x: 1
    if x["Count of Democratic Voters"] > x["Count of Republican Voters"]
    else 0,
    axis=1,
)
```

```python
df["Orientation2"] = df.apply(
    lambda x: x["Count of Democratic Voters"] / x["Total Count of All Voters"]
    if x["Count of Democratic Voters"] > x["Count of Republican Voters"]
    else -x["Count of Republican Voters"] / x["Total Count of All Voters"],
    axis=1,
)
```
